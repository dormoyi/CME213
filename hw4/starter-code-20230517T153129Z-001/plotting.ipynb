```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. For order = 4, plot the bandwidth for the 2 (or 3) different algorithms.
# plot at the 5 locations of grid size: 256, 512, 1024, 2048, 4096 with the labels
grid_size = [256, 512, 1024, 2048, 4096]
labels = ["256x256", "512x512", "1024x1024", "2048x2048", "4096x4096"]
algo1 = [475.836, 1190.61, 1565.88, 1875.03, 1968.88]
algo2 = [401.982, 1329.44, 1621.54, 1860.48, 1931.82]
algo3 = [362.96, 1193.21,  2041.11, 1821.99, 1908.55]

plt.figure(figsize=(15, 5))
plt.plot(grid_size, algo1, label='Global')
plt.plot(grid_size, algo2, label='Block')
plt.plot(grid_size, algo3, label='Shared')
# plot the labels on the x-axis
plt.xticks(grid_size, labels)
#plt.plot(grid_size, algo3, label='Algorithm 3')
plt.xlabel('Grid Size')
plt.ylabel('Bandwidth (GB/s)')
plt.legend()
plt.title('Bandwidth vs Grid Size for Order = 4')
plt.savefig('bandwidth_order4.png')
```
