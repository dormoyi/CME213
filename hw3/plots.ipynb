```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv('q1.4.csv', header=None, sep=' ')
y1 = list(df1.iloc[0])[:-1]
x1 = [i for i in range(32, 1025, 32)]
plt.plot(x1, y1, label='q1.4')
plt.xlabel('Number of threads')
plt.ylabel('Performance (TFlops/sec)')
plt.title('Question 1.4: vary number of threads for a small number of blocks')
plt.savefig('q1.4.png')
```
